```python
from PIL import Image
import numpy as np
import utils

from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}

img{
    width: auto;
}

</style>
""")
```

```python
# First we must load the image we will use
img = np.asarray(Image.open('myramen.jpg'))
Image.fromarray(img)
```

```python
# Next stape: Reshaping to apply kmeans (We need 3 columns = "RGB")
x, y, z = img.shape
img = img.reshape(x*y, z)
```

```python
# We apply our kmeans algorithm using k=8 (8 groups)
c,l = utils.kmean(img, k = 8)

print('Shape of centroids: ', c.shape)
print('Shape of labels: ', l.shape)
```

```python
# Finally we reshape the array back to its previous dimensions and convert it to an image
img2 = c[l].reshape(x,y,z)
Image.fromarray(img2.astype(np.uint8))
```

```python
# Let's now try using k=16 (16 groups)
c,l = utils.kmean(img, k = 16)

print('Shape of centroids: ', c.shape)
print('Shape of labels: ', l.shape)
```

```python
img2 = c[l].reshape(x,y,z)
Image.fromarray(img2.astype(np.uint8))
```
